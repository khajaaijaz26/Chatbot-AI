```python
# chatbot.py

def simple_chatbot(user_query):
    if user_query == "What is the total revenue?":
        return "The total revenue is $1,200,000."
    elif user_query == "How has net income changed over the last year?":
        return "The net income has increased by $150,000 over the last year."
    elif user_query == "What is the current profit margin?":
        return "The current profit margin is 15%."
    elif user_query == "What are the operating expenses?":
        return "The operating expenses are $500,000."
    elif user_query == "What is the total debt?":
        return "The total debt is $800,000."
    else:
        return "Sorry, I can only provide information on predefined queries."

# Run chatbot interaction in command-line
if __name__ == "__main__":
    while True:
        user_input = input("Ask a question: ")
        if user_input.lower() in ["exit", "quit"]:
            print("Goodbye!")
            break
        response = simple_chatbot(user_input)
        print(response)
```
